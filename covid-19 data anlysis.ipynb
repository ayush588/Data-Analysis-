```python
# In this part we have imported the important tools for working .
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
%matplotlib inline 

import plotly
import plotly.express as px
import plotly.graph_objects as go
plt.rcParams['figure.figsize']=17,8

import cufflinks as cf
import plotly.offline as pyo
from plotly.offline import init_notebook_mode,plot,iplot

import folium
```

```python
# dataset
df = pd.read_excel(r'C:\Users\user\Downloads\Covid cases in India.xlsx')
```

```python
df.head()
```

```python
# we droped the S.No. pemanently
df.drop(['S. No.'],axis=1,inplace=True)
```

```python
df.head()
```

```python
# We had made the new column of Total Cases in India 
df['Total Cases']=df['Total Confirmed cases (Indian National)'] + df['Total Confirmed cases ( Foreign National )']
```

```python
df.head()
```

```python
# Just to show (it is not recommended)
total_cases_overall = df['Total Cases'].sum()
print('The total number of cases in India ',total_cases_overall)
```

```python
df.head()
```

```python
# again new column of Active cases in India 
df['Active cases'] = df['Total Cases'] - df['Cured']+df['Death']
```

```python
df.head()
```

```python
# Given Style to understand better.
df.style.background_gradient(cmap='Reds')
```

```python
# Sorted the active cases and given style 
Total_Active_Cases = df.groupby('Name of State / UT')['Active cases'].sum().sort_values(ascending=False).to_frame()
Total_Active_Cases.style.background_gradient(cmap='Reds')
```

```python
# taken data set to plot on a map
Indian_cord = pd.read_excel(r'C:\Users\user\Downloads\Indian Coordinates.xlsx')
```

```python
Indian_cord.head()
```

```python
df_full=pd.merge(Indian_cord,df,on='Name of State / UT')
df_full
```

```python
# locating on the map
map = folium.Map(location=[20,70],zoom_start=4,tiles='Stamenterrain')
for lat,long,value,name in zip(df_full['Latitude'],df_full['Longitude'],
                               df_full['Total Cases'],
                               df_full['Name of State / UT']):
    folium.CircleMarker([lat,long],radius=value*0.8,
                        popup=('<strong>State</strong>:'+str(name).capitalize()
                               +'<br>''<strong> Total Cases </strong>: '+str(value)+'<br'),
                        color='red',fill_color='red',fill_opacity=0.3).add_to(map)
```

```python
# to make subplots we will import this.
from plotly.subplots import  make_subplots
```

```python
# given style for subplots 
fig=make_subplots(
    rows=2,cols=2,specs=[[{'secondary_y':True},{'secondary_y':True}],[{'secondary_y':True},{'secondary_y':True}]],
    subplot_titles=('S.Korea','Italy','India','Wuhan'))

# given data ,style and other parameters given below 
fig.add_trace(go.Bar(x=dbd_Korea['Date'],y=dbd_Korea['Total Cases'],
                    marker=dict(color=dbd_Korea['Total Cases'],coloraxis='coloraxis')),1,1)
fig.add_trace(go.Bar(x=dbd_Italy['Date'],y=dbd_Italy['Total Cases'],
                    marker=dict(color=dbd_Italy['Total Cases'],coloraxis='coloraxis')),1,2)
fig.add_trace(go.Bar(x=dbd_India['Date'],y=dbd_India['Total Cases'],
                    marker=dict(color=dbd_India['Total Cases'],coloraxis='coloraxis')),2,1)
fig.add_trace(go.Bar(x=dbd_Wuhan['Date'],y=dbd_Wuhan['Total Cases'],
                    marker=dict(color=dbd_Wuhan['Total Cases'],coloraxis='coloraxis')),2,2)

fig.update_layout(coloraxis=dict(colorscale='Bluered_r'),showlegend=False,title_text='Total cases in 4 countries.')

fig.update_layout(plot_bgcolor='rgb(230,230,230)')
```

```python
df.rename(columns={'ObservationDate':'Date','Country/Region':'Country'},inplace=True)
```

```python
df.query('Country=="UK"')
```

```python
df.groupby('Date').sum()
```

```python
confirmed = df.groupby('Date').sum()['Confirmed'].reset_index()
deaths= df.groupby('Date').sum()['Deaths'].reset_index()
Recover = df.groupby('Date').sum()['Recovered'].reset_index()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=confirmed['Date'],y=confirmed['Confirmed'],mode='lines+markers',name='Confirmed',line=dict(color='blue',width=2)))

fig.add_trace(go.Scatter(x=deaths['Date'],y=deaths['Deaths'],mode='lines+markers',name='Deaths',line=dict(color='red',width=2)))

fig.add_trace(go.Scatter(x=Recover['Date'],y=Recover['Recovered'],mode='lines+markers',name='Recovered',line=dict(color='green',width=2)))
```
